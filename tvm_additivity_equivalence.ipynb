As we have demonstrated, finding present and future values involves moving amounts of money to differ points on a time line. 

These opeations are possible because present value and future value are equivalent measures in separate time. 
image.png

```python

## The Equivalence of Present and Future Values
##define variables 
Time = [-1, -2, -3, -4, -5]
cash_flow = 1000
r = .05

##present value in array
present_value = []

for t in Time:
    pv = cash_flow / ((1 + r) ** (-t))
    present_value.append(pv)

print(present_value)

##future value variavles

TimeFV = [4, 3, 2, 1, 0]

future_value = []

for t in TimeFV:
  fv = cash_flow * (1 + r) ** (t)
  future_value.append(fv)

print(future_value)
```

 A lump sum can actually generate an annuity. If we place a lump sum in an account that earns the stated interest rate for all periods, we can generate an annuity that is equivalent to the lump sum. Amortized loans, such as mortgages and car loans, are examples of this principle.
