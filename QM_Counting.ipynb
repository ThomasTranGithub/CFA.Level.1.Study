**Multiplication rule for counting**

If one task can be done in n_1 ways, and a second task, given the first, can be done in n_2 ways, and a third task, given the first two task can be done in n_3 ways, and so on for k tasks, then the number of ways the k tasks can be done is n_1 * n_2 * n_3....n_k.
