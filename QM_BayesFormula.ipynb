**Bayes' Formula**

a rational method for adjusting our viewpoints

Given a set of prior probabilities for an event of interest, if you receive new information, the ruel for update your porbability of the event is


updated probability of event given new information is equal to

probability of the new information given event divided by the unconditional probability of the new event

multiplied by the prior probability of the event.

P(Event|Information) =

P(Information|Event)/P(Information) * P(Event)
